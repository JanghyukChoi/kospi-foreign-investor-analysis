```python
from selenium import webdriver
import sys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import FinanceDataReader as fdr
import pandas as pd
from datetime import datetime
import mojito
import time
import json
import asyncio
import requests
from datetime import timedelta
import telegram
from dotenv import load_dotenv
import os

APP_KEY = os.getenv("APP_KEY")
APP_SECRET = os.getenv("APP_SECRET")
acc_no = os.getenv("acc_no")
# 실전투자
URL_BASE = os.getenv("URL_BASE")
# Selenium 웹드라이버 설정

processed_companies = []  # Using list instead of set


TOKEN = os.getenv("APP_SECRET")
MYCHAT_ID = os.getenv("MYCHAT_ID")
token = os.getenv("token")


def get_access_token():
    """토큰 발급"""
    headers = {"content-type": "application/json"}
    body = {"grant_type": "client_credentials",
            "appkey": APP_KEY,
            "appsecret": APP_SECRET}
    PATH = "oauth2/tokenP"
    URL = f"{URL_BASE}/{PATH}"
    res = requests.post(URL, headers=headers, data=json.dumps(body))
    print(res.json())
    ACCESS_TOKEN = res.json()["access_token"]
    return ACCESS_TOKEN

ACCESS_TOKEN = get_access_token()

# API 요청 정보
url = "https://openapi.koreainvestment.com:9443/uapi/domestic-stock/v1/quotations/inquire-price"
headers = {
    "content-type": "application/json; charset=utf-8",
    "authorization": f"Bearer {ACCESS_TOKEN}",
    "appkey": APP_KEY,
    "appsecret": APP_SECRET,
    "tr_id": "FHKST01010100"  # 개인투자자용 거래 ID
}



# 요청 파라미터
params = {
    "fid_cond_mrkt_div_code": "J",  # 코스피
    "fid_input_iscd": "005930"      # 삼성전자 종목 코드
}

# API 호출
response = requests.get(url, headers=headers, params=params)
print(response.json())

```
