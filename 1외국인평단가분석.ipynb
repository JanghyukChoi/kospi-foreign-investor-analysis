## 외국인 기간별 평균 단가 및 평균 수익률

```python
import pandas as pd
from pykrx import stock
from datetime import datetime, timedelta

# 종목 코드 설정
ticker = "005930"  # 삼성전자

# 오늘 날짜 기준으로 최근 1개월, 3개월, 6개월, 1년 날짜 계산
today = datetime.today()
date_ranges = {
    "3 Days": (today - timedelta(days=3)).strftime("%Y%m%d"),
    "1 Week": (today - timedelta(days=7)).strftime("%Y%m%d"),
    "2 Week": (today - timedelta(days=14)).strftime("%Y%m%d"),
    "1 Month": (today - timedelta(days=30)).strftime("%Y%m%d"),
    "3 Months": (today - timedelta(days=90)).strftime("%Y%m%d"),
    "6 Months": (today - timedelta(days=180)).strftime("%Y%m%d"),
    "1 Year": (today - timedelta(days=365)).strftime("%Y%m%d"),
}

# 결과 저장 딕셔너리
average_prices = {}
net_trading_values = {}  # 순거래대금 저장용 딕셔너리

# 현재가 가져오기
end_date = today.strftime("%Y%m%d")
current_price = stock.get_market_ohlcv_by_date(end_date, end_date, ticker)['종가'].iloc[0]

# 각 기간에 대해 외국인 평균 매수 단가 계산
for period, start_date in date_ranges.items():
    end_date = today.strftime("%Y%m%d")
    
    # 거래 데이터 가져오기
    value_data = stock.get_market_trading_value_by_investor(start_date, end_date, ticker)
    volume_data = stock.get_market_trading_volume_by_investor(start_date, end_date, ticker)
    
    # 외국인 데이터 필터링
    foreign_value = value_data.loc["외국인"]["매수"]
    foreign_volume = volume_data.loc["외국인"]["매수"]
    foreign_net_value = value_data.loc["외국인"]["순매수"]

    # 평균 매수 단가 계산
    foreign_average_price = (foreign_value / foreign_volume)

    # 결과 저장
    average_prices[period] = foreign_average_price
    net_trading_values[period] = foreign_net_value

# 결과 출력
for period in date_ranges.keys():
    average_price = average_prices[period]
    profit_or_loss = ((current_price - average_price) / average_price) * 100
    print(f"{period} - 평균단가 : {average_price:.2f} KRW  누적 순매수 거래대금 : {net_trading_values[period]:.2f} KRW  수익률 : {profit_or_loss:.2f}%")
```

## 외국인 평균 매수 단가 및 누적 순매수 거래대금

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager as fm

# 종목명 변수 설정
stock_name = "삼성전자"

# 한글 폰트 설정 (맑은 고딕)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 마이너스(-) 폰트 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# period 목록
periods = list(average_prices.keys())

# DataFrame 생성
df = pd.DataFrame({
    'Average_Price': [average_prices[p] for p in periods],
    'Net_Trading_Value': [net_trading_values[p] for p in periods]
})
df['Period'] = periods

# 현재가와 수익률 컬럼 추가
df['Profit_or_Loss'] = ((current_price - df['Average_Price']) / df['Average_Price']) * 100

# 원하는 순서 (1 Year가 왼쪽, 3 Days가 오른쪽)
period_order = ["1 Year", "6 Months", "3 Months", "1 Month", "2 Week", "1 Week", "3 Days"]

# Period 열을 카테고리로 변환하여 순서 지정 후 정렬
df['Period'] = pd.Categorical(df['Period'], categories=period_order, ordered=True)
df.sort_values('Period', inplace=True)

# 막대 그래프 + 라인 그래프 (보조 y축)
fig, ax1 = plt.subplots(figsize=(10,6))

# 1차 축: 평균 매수 단가 막대 그래프
ax1.bar(df['Period'], df['Average_Price'], color='skyblue', label='외국인 평균 매수 단가')
ax1.set_xlabel('Period')
ax1.set_ylabel('평균 매수 단가 (KRW)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# 2차 축: 누적 순매수 거래대금 라인 그래프
ax2 = ax1.twinx()
ax2.plot(df['Period'], df['Net_Trading_Value'], marker='o', color='green', label='누적 순매수 거래대금')
ax2.set_ylabel('누적 순매수 거래대금 (KRW)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# 제목에 종목명 추가
plt.title(f"{stock_name} 외국인 평균 매수 단가 및 누적 순매수 거래대금")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,6))

# x축이 문자열일 때 수직선(vlines) 등을 깔끔하게 그리기 위해, 0부터 len(df)-1까지 번호를 붙여줍니다.
x_positions = np.arange(len(df))

# (A) 평균 매수가 라인 그래프
ax.plot(x_positions,
        df['Average_Price'],
        marker='o',
        label='외국인의 현재 평균단가 (~ 달 전부터 지금까지의 데이터 고려)',
        color='blue')

# (B) 현재가 수평선
ax.axhline(y=current_price, color='orange', linestyle='--', label='현재가')

# (C) 각 점마다 현재가와의 수직선 + 수익률 표시
for i, row in enumerate(df.itertuples()):
    avg_price = row.Average_Price
    x = x_positions[i]
    
    # 수직선: 평균단가 ~ 현재가
    ax.vlines(x, 
              ymin=min(avg_price, current_price),
              ymax=max(avg_price, current_price),
              color='red', 
              linestyle='-', 
              linewidth=2)
    
    # 손익률 계산
    pnl = ((current_price - avg_price) / avg_price) * 100
    
    # 수직선 중간 지점에 텍스트 붙이기
    mid_y = (avg_price + current_price) / 2
    ax.text(x,
            mid_y,
            f'{pnl:+.2f}%',
            color='black',
            ha='center',
            va='center',
            fontsize=9,
            fontweight='bold')

# (D) X축에 Period 라벨 달기
ax.set_xticks(x_positions)
ax.set_xticklabels(df['Period'])

# 레이아웃 꾸미기
ax.set_xlabel('Period')
ax.set_ylabel('KRW')
ax.set_title('Avg Price vs. Current Price')
ax.legend()
plt.tight_layout()
plt.show()
```
